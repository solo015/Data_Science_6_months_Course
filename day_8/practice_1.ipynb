```python
# import library
import pandas as pd
```

```python
# Loading the data set
df = pd.read_csv("pk_data.csv")
```

# loc()

```python
# loc() to select rows and columns
df.loc[10:100,"DIVISION":"POPULATION 1998 (RURAL)"]
# this operations selected the columns from "Division" to "Population 1998 (Rural)" and rows from 10th to 100th records.
```

```python
df.loc[df['FEMALE (RURAL)'] > 10, ['FEMALE (RURAL)','DIVISION','DISTRICT','PROVINCE']]

# select rows meeting logical conditions like df['FEMALE (RURAL)'] > 10 
# and only the specific columns like ['FEMALE (RURAL)','DIVISION','DISTRICT','PROVINCE']
```

# iloc()

```python
df.iloc[ 10:20] # 10 to 20 range is index values
# here select the rows 10-20
```

```python
# select columns in positions 3,4,6
print( df.iloc[:, [3,4,6]] ) # columns : 3,4,6 and rows ':' are all.



# select columns in positions 3,5,7 and rows from 100 to 150
print( df.iloc[100: 150, [3,5,7]])
# columns : 3,5,7
# rows : from 100 to 150
```
